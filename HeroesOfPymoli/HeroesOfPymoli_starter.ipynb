```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_player_count = purchase_data.shape[0]
print(total_player_count)
player_list = pd.Series(purchase_data['SN'])
unique_players = player_list.unique()
number_unique_players = len(unique_players)
print(number_unique_players)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Rename the dataframe so I remember what it's called
df = purchase_data
number_purchases_per_player = df['SN'].value_counts()
number_purchases_per_player.head(50)
#No real extreme outliers

#Number of unique items
items = pd.Series(df['Item Name']).unique()
number_items = len(items)
print(f'There are {number_items} unique purchased items.')

#average purchase price
#let's check out the distribution first
df['Price'].describe()
#looks relatively normal, just use the mean
avg_price = round(df['Price'].mean(),2)
print(f'The average price is ${avg_price}.')

#Total purchases
total_purchases = total_player_count
print(f'There were {total_purchases} total purchases.')

#Total Revenue
revenue = df['Price'].sum()
print(f'We made ${revenue} off of these suckers! HA!')

#Combine into a dataframe
#purchase_analysis = pd.DataFrame([number_items,avg_price,total_purchases,revenue],index=['Number of Unique Items Purchased','Average Price of Purchases','Total Purchases','Total Revenue'])
purchase_analysis = pd.DataFrame({'Number of Unique Items Purchased': [number_items],'Average Price of Purchases': [avg_price],'Total Purchases': [total_purchases],'Total Revenue': [revenue]})
purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players

```python
male_count = len(df.loc[df['Gender']=='Male','SN'].unique())
male_per = round((male_count/total_player_count)*100,2)
female_count = len(df.loc[df['Gender']=='Female','SN'].unique())
female_per = round((female_count/total_player_count)*100,2)
other_count = len(df.loc[df['Gender']=='Other / Non-Disclosed','SN'].unique())
other_per = round((other_count/total_player_count)*100,2)

gender_demos = pd.DataFrame({'Gender':['Male','Female','Other/Non-disclosed'],
                             'Count':[male_count,female_count,other_count],
                             'Percentage':[male_per,female_per,other_per]})
gender_demos = gender_demos.set_index('Gender')
gender_demos
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase Count
male_purchases = df['Gender'].value_counts()[0]
female_purchases = df['Gender'].value_counts()[1]
other_purchases = df['Gender'].value_counts()[2]
#average purchase price
male_avg_price = round(df.loc[df['Gender']=='Male','Price'].mean(),2)
female_avg_price = round(df.loc[df['Gender']=='Female','Price'].mean(),2)
other_avg_price = round(df.loc[df['Gender']=='Other / Non-Disclosed','Price'].mean(),2)
#total purchase value
male_purchase_value = df.loc[df['Gender']=='Male','Price'].sum()
female_purchase_value = df.loc[df['Gender']=='Female','Price'].sum()
other_purchase_value = df.loc[df['Gender']=='Other / Non-Disclosed','Price'].sum()
#average purchase value per person by gender
male_avg_purchase_value = male_purchase_value/male_count
female_avg_purchase_value = female_purchase_value/female_count
other_avg_purchase_value = other_purchase_value/other_count

gender_purchases = pd.DataFrame({'Gender':['Male','Female','Other/Non-disclosed'],
                                 'Purchase Count':[male_purchases,female_purchases,other_purchases],
                                 'Average Purchase Price':[male_avg_price,female_avg_price,other_avg_price],
                                 'Total Purchase Value':[male_purchase_value,female_purchase_value,other_purchase_value],
                                 'Average Purchase Value':[male_avg_purchase_value,female_avg_purchase_value,other_avg_purchase_value]})
gender_purchases = gender_purchases.set_index('Gender')
gender_purchases
```
